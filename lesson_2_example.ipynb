## Read the image

### Import the library

```python
import cv2
```

### Use cv2.imread() to read an image file with BGR format.


```python
# cv2.imread(const String &	filename, int	flags = IMREAD_COLOR_BGR)
# The loaded image must be BGR color when we call the cv2.imread() without changing the flags

image = cv2.imread('/content/fishball_stall.png')
```

### To show the image is an multiple dimension array

```python
# type() will get the parameter type
# we will show the image type
# it should be return numpy.ndarray, nd means n of dimension
print(type(image))
```

### * Extract the image properties from image file

```python
height, width, channels = image.shape
# image.shape will return 3 array
print(width, height, channels)


# Channel should be contains RGB value and each of the value should be 0-255
# R (Red) / G (Green) / B (Blue)
# Note that: when we load the image from OpenCV it should be BGR.


# image.shape = [
#     # first level of array is height array (total 350)
#     [
#         # second level of array is width array (total 515)
#         [0,255,0], # third level of array is channel array [B,G,R] (total 3)
#         [112,255,123],
#          ...total 515
#     ],
#     [],
#     [],
#     ...total 350
# ]


# Refnerence: https://docs.opencv.org/4.x/d3/df2/tutorial_py_basic_ops.html
```

### Converting from BGR to RGB.


```python
# cv2.cvtColor(Array source, int code)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # To convert the color into the RGB model

# cv2.COLOR_BGR2RGB is a constant number that means BGR to RGB
```

### Use matplotlib to display an image in Google colab.
#### If you are using local python then you can use cv2.imshow()

```python
import matplotlib.pyplot as plt

plt.imshow(image) # Put the image into the matplotlib
plt.axis('off')  # hide the axis
plt.show() # Show the image
```

## Manipulate the image


### Create the image


```python
import cv2
import numpy as np # numpy bundled in opencv you don't need to install again
import random
import matplotlib.pyplot as plt

width = 300
height = 300

# create the empty image
img = np.empty((height, width, 3), np.uint8) # unit means positive integer with 8 bits

# generate the random color image
for row in range(height):  # this row array is height
    for col in range(img.shape[1]):  # this col array is width
        img[row][col] = [ # this array is storing RGB value
            random.randint(0, 255), # R
            random.randint(0, 255), # G
            random.randint(0, 255)  # B
        ]
plt.imshow(img) # Put the image into the matplotlib
plt.axis('off')  # hide the axis
plt.show() # Show the image
```

### Resize image resolution

```python
import cv2
import random
import matplotlib.pyplot as plt
# cv2.resize(image_source, (width, height))

resized_image = cv2.resize(image, (100, 100)) # resize the image to 100px x 100px

# if you want to scale the image without hard code then you can
# resized_image = cv2.resize(image, (0, 0), fx=0.5, fy=0.5) # reduce the entire image by half

plt.imshow(resized_image) # Put the image into the matplotlib
plt.axis('off')  # hide the axis
plt.show() # Show the image
```

### Adjusting the brightness

```python
import cv2
import random
import matplotlib.pyplot as plt

img = cv2.imread('/content/fishball_stall.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# adjust the image brighter
bright_image = cv2.convertScaleAbs(img, beta=100)

plt.imshow(bright_image) # Put the image into the matplotlib
plt.axis('off')  # hide the axis
plt.show() # Show the image
```

### Adjusting the contrast

```python
import cv2
import random
import matplotlib.pyplot as plt

img = cv2.imread('/content/fishball_stall.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# adjust the image hight contrast
contrast_image = cv2.convertScaleAbs(image, alpha=1.8)

plt.imshow(contrast_image) # Put the image into the matplotlib
plt.axis('off')  # hide the axis
plt.show() # Show the image
```

## Basic Filtering

### Blur

```python
import cv2
import random
import matplotlib.pyplot as plt

img = cv2.imread('/content/fishball_stall.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# make the image blur
blurred_image = cv2.GaussianBlur(image, (9, 9), 0)

plt.imshow(blurred_image) # Put the image into the matplotlib
plt.axis('off')  # hide the axis
plt.show() # Show the image
```

### Canny

```python
import cv2
import random
import matplotlib.pyplot as plt

img = cv2.imread('/content/fishball_stall.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# make the image blur
edges = cv2.Canny(image, 10, 100)

plt.imshow(edges) # Put the image into the matplotlib
plt.axis('off')  # hide the axis
plt.show() # Show the image
```
